# Feature Engineering

```python
df["pdays"].describe()
# Most customers have not been contacted before
# Split into two columns - contacted before or not and days since last contact

df["previous_contact"] = df["pdays"].apply(lambda x: 0 if x == -1 else 1)
df["days_since_last_contact"] = df["pdays"].apply(lambda x: x if x != -1 else 0)
df = df.drop("pdays", axis=1)
```

# Variable Encoding

```python
# Encode binary and categorical variables for modelling
binary_cols = ["default", "housing", "loan", "y"]
for col in binary_cols:
    df[col] = df[col].map({"yes": 1, "no": 0}).astype(int)
```

```python
categorical_cols = ["job", "marital", "education", "contact", "month", "poutcome"]
df = pd.get_dummies(
    df,
    columns=categorical_cols,
    prefix=categorical_cols,
    drop_first=True,
    dtype=int,
)
```
